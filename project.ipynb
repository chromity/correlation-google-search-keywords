# Project - Correlation Analysis in Python

## Scenario/Background

Correlation is an important statistical relationship that can indicate whether the variable values are linearly related.

In this project we used Google Search Data Analytics of the Philippines with specific keywords. In our case, we used the keywords "tesda", "hiring philippines", and "hiring abroad". We then used correlation to analyze the relationship between the data.

The search data gathered are weekly series.

### Calculate Correlation

We used different correlation methods to calculate the correlation between the keywords. 

```python
tesdaHiringPH.corr(method='pearson')

```

```python
tesdaHiringPH.corr(method='kendall')
```

```python
tesdaHiringPH.corr(method='spearman')
```

```python
tesdaHiringNonPH.corr(method='pearson')
```

```python
tesdaHiringNonPH.corr(method='kendall')
```

```python
tesdaHiringNonPH.corr(method='spearman')
```

In the results, we discovered that the pearson method gives the highest correlation from the three.
